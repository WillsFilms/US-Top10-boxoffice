Set up the analysis by importing relevant packages and the data (csv)

```python
#plot of Twilight budgets over time

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.rcsetup 
from matplotlib import rcParams #allow parameters
import seaborn as sns

Twilight_movies = pd.read_csv("C:\\Users\\HP\\Documents\\Data Analysis\\Projects\\Films\\Box Office\\Box Office Top 10\\twilight.csv")
```

Assign the variables

```python
budget_values = Twilight_movies['Budget ($)']
films = Twilight_movies['Film']
rank = Twilight_movies['Rank']
```

calculate the median budget value and assign to a variable

```python
median_budget = np.median(budget_values)
print(median_budget) #test 
```

Plot the data in a line plot

```python
matplotlib.rcParams["axes.formatter.limits"] = (-99, 99) #removes scientific labelling

fig, ax = plt.subplots()

ax.plot(films, budget_values, color='#44546A',
        linewidth=4, markersize=8)
ax.set_title('Twilight film budgets ($)',  
             size = 18,
          pad = 10)
ax.set_ylim(0, 140000000)
ax.set_yticklabels(['{:,}'.format(int(x))for x in ax.get_yticks().tolist()]) #adds commas to large numbers

ax.axhline(median_budget, color='r', linestyle='--',
           label='Median Budget') #plot average budget line

ax.legend(fontsize='12')

sns.despine()

plt.tight_layout()
plt.show()
```
